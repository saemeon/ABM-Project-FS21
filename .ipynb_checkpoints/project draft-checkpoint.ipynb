```python
n = 50
m = 1
s = 1
shifts = [0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,1,1.25,1.5,1.75,2,2.5,3,3.5,4,5,8]

for Mshift in tqdm(shifts): 
    try:
        ResultsAna[Mshift]
    except:
        ResultsAna[Mshift]={}
    M  = np.array(n*[np.linspace(0 , m, n)])
    M0 = M+Mshift
    Sshift = 0
    S = np.array(n*[np.linspace(s,0.001,n)]).T
    S0 = S+Sshift
    
    #con
    X_list = []
    X = norm.cdf(0,M,S)
    X_ = X.copy()
    for i in range(50):
        X_list.append(X)
        X_1 = X.copy()
        X  = norm.cdf(X,M,S) - norm.cdf((1-X),M0,S0)*X
        if np.array_equal(X_1, X):
            break
    ResultsAna[Mshift]["con"]=np.array(X_list)
    
    #red
    X_list = []
    X = norm.cdf(0,M,S)
    X_ = X.copy()
    for i in range(100):
        X_list.append(X)
        X_1 = X.copy()
        X  = norm.cdf(X,M,S) - norm.cdf((1-X)/X,M0,S0)*X
        if np.array_equal(X_1, X):
            break
    ResultsAna[Mshift]["red"]=np.array(X_list)
```
